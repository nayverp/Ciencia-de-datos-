#ejercicios del 1 al 5

```python
print("Calcular area de un triangulo")
base = float(input("Ingrese la base: "))
altura = float(input("Ingrese la altura: "))
area = (base * altura) / 2
print("El area es:", area)
```

```python
print("Convertir Celsius a Fahrenheit")
c = float(input("Ingrese grados Celsius: "))
f = (c * 9/5) + 32
print("Temperatura en Fahrenheit:", f)
```

```python
print("Calcular promedio de tres numeros")
n1 = float(input("Ingrese el primer numero: "))
n2 = float(input("Ingrese el segundo numero: "))
n3 = float(input("Ingrese el tercer numero: "))
promedio = (n1 + n2 + n3) / 3
print("El promedio es:", promedio)
```

```python
print("Calcular perimetro y area de un circulo")
radio = float(input("Ingrese el radio: "))
perimetro = 2 * 3.1416 * radio
area = 3.1416 * radio * radio
print("Perimetro:", perimetro)
print("Area:", area)
```

```python
print("Resolver ecuacion cuadratica")
a = float(input("Ingrese a: "))
b = float(input("Ingrese b: "))
c = float(input("Ingrese c: "))
discriminante = (b*b) - (4*a*c)
x1 = (-b + discriminante**0.5) / (2*a)
x2 = (-b - discriminante**0.5) / (2*a)
print("Solucion x1:", x1)
print("Solucion x2:", x2)
```

```python
print("Determinar si un numero es par o impar")
numero = int(input("Ingrese un numero: "))
if numero % 2 == 0:
    print("El numero es par")
else:
    print("El numero es impar")
```

```python
print("Determinar si un año es bisiesto")
añio = int(input("Ingrese el año: "))
if (añio % 4 == 0 and añio % 100 != 0) or (añio % 400 == 0):
    print("Es un año bisiesto")
else:
    print("No es un año bisiesto")
```

```python
print("Prioridad por edad")
nombre = input("Ingrese su nombre: ")
edad = int(input("Ingrese su edad: "))
if edad >= 70:
    print(nombre, "tiene prioridad en la fila")
else:
    print(nombre, "no tiene prioridad")
```

```python
print("Comparar dos numeros")
n1 = float(input("Ingrese el primer numero: "))
n2 = float(input("Ingrese el segundo numero: "))
if n1 > n2:
    print("El mayor es:", n1)
else:
    print("El mayor es:", n2)
```

```python
print("Determinar tipo de triangulo")
a = float(input("Ingrese lado 1: "))
b = float(input("Ingrese lado 2: "))
c = float(input("Ingrese lado 3: "))
if a == b and b == c:
    print("Triangulo equilatero")
else:
    if a == b or a == c or b == c:
        print("Triangulo isosceles")
    else:
        print("Triangulo escaleno")
```

```python
print("Calculo de IMC")
nombre = input("Ingrese su nombre: ")
peso = float(input("Ingrese su peso (kg): "))
altura = float(input("Ingrese su altura (cm): "))

altura = altura / 100

imc = peso / (altura * altura)

if imc < 18.5:
    categoria = "Bajo peso"
    mensaje = "Debe mejorar su alimentacion"
elif imc < 25:
    categoria = "Normal"
    mensaje = "Mantenga habitos saludables"
elif imc < 30:
    categoria = "Sobrepeso"
    mensaje = "Se recomienda ejercicio"
else:
    categoria = "Obesidad"
    mensaje = "Consulte a un especialista"

print(nombre, "su IMC es:", imc)
print("Categoria:", categoria)
print("Recomendacion:", mensaje)
```

```python
print("Calculadora simple")
n1 = float(input("Ingrese primer numero: "))
n2 = float(input("Ingrese segundo numero: "))
op = input("Operacion (+, -, *, /): ")
if op == "+":
    resultado = n1 + n2
elif op == "-":
    resultado = n1 - n2
elif op == "*":
    resultado = n1 * n2
elif op == "/":
    resultado = n1 / n2
else:
    resultado = "Operacion no valida"
print("Resultado:", resultado)
```

```python
print("Calificacion en letras")
nota = int(input("Ingrese la nota (0-100): "))
if nota >= 90:
    letra = "A"
elif nota >= 80:
    letra = "B"
elif nota >= 70:
    letra = "C"
elif nota >= 60:
    letra = "D"
else:
    letra = "F"
print("Calificacion:", letra)
```

```python
print("Dia de la semana")
dia = int(input("Ingrese numero (1-7): "))
if dia == 1:
    nombre = "Lunes"
elif dia == 2:
    nombre = "Martes"
elif dia == 3:
    nombre = "Miercoles"
elif dia == 4:
    nombre = "Jueves"
elif dia == 5:
    nombre = "Viernes"
elif dia == 6:
    nombre = "Sabado"
elif dia == 7:
    nombre = "Domingo"
else:
    nombre = "Numero invalido"
print("Dia:", nombre)
```

```python
print("Validar triangulo")
a = float(input("Ingrese lado 1: "))
b = float(input("Ingrese lado 2: "))
c = float(input("Ingrese lado 3: "))
if a + b > c and a + c > b and b + c > a:
    if a == b and b == c:
        tipo = "Equilatero"
    elif a == b or a == c or b == c:
        tipo = "Isosceles"
    else:
        tipo = "Escaleno"
    print("Si forma triangulo:", tipo)
else:
    print("No forma un triangulo")
```
